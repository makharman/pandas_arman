```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt #для визуализации
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)# создаем диаграмму
fig.set_size_inches(20,4)#устанавливаем размер диаграммы
x = [1,2,3,4]
y = [4,3,2,1]
y1 = [4,3,2,2]

ax1.plot(x, y,"g--") #показываем оси
ax1.set_title("UFO") # название диаграммы

ax2.plot(x, y1) #показываем оси
ax2.set_title("UFO2") # название диаграммы

plt.show()
```

## ТОП 5 города

```python
cities = (df2[["city","datetime"]]
 .groupby("city")
 .count()
 .rename(columns={"datetime":"cnt"})
 .sort_values("cnt",ascending=False)
).head()
```

```python
cities.plot.bar()
plt.title("top cities")
```

## Время наблюдении

```python
import seaborn as sns
```

```python
year_group = (df2[["year","datetime"]]
 .groupby("year")
 .count()
 .rename(columns={"datetime":"cnt"})
).loc[:2013]
```

```python
year_group
```

```python
plt.figure(figsize=(20,5))
sns.lineplot(x=year_group.index, y=year_group.cnt, data=year_group)
plt.show()
```

```python
df2.head(2)
```

```python
df2.loc[:,"weekday"] = df2["datetime"].dt.dayofweek
```

```python
df2.head()
```

```python
df2["weekday"].value_counts()
```

```python
weekday_group = (df2[["weekday","datetime"]]
 .groupby("weekday")
 .count()
 .rename(columns={"datetime":"cnt"})
)
```

```python
weekday_group
```

```python
sns.barplot(x=weekday_group.index, y=weekday_group["cnt"], data = weekday_group, palette="viridis")
```

```python
hour_group = (df2[["hour","datetime"]]
 .groupby("hour")
 .count()
 .rename(columns={"datetime":"cnt"})
).loc[:2013]
```

```python
plt.figure(figsize=(20,5))
plt.title("Visiting hours")
sns.barplot(x=hour_group.index, y=hour_group["cnt"], data = hour_group, palette="viridis")
hour_group
```
